```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img0=cv2.imread('0.jpg',0)
plt.imshow(img0,cmap='gray')
```

```python
img0 = cv2.medianBlur(img0,5)
img0_circles=cv2.HoughCircles(img0,cv2.HOUGH_GRADIENT,1,50,param1=40,param2=40,minRadius=0,maxRadius=0)
img0_circles=np.uint16(np.around(img0_circles))
#print(img0_circles)
```

```python
circle_img0=cv2.cvtColor(img0,cv2.COLOR_GRAY2BGR)
```

```python
img0_counter=0
for i in img0_circles[0,:]:
    cv2.circle(circle_img0,(i[0],i[1]),i[2],(255,90,170),3)
    img0_counter+=1

cv2.putText(circle_img0,str(img0_counter),(50,50),cv2.FONT_HERSHEY_SIMPLEX,1,(255,0,255),5,cv2.LINE_AA)
plt.imshow(circle_img0,cmap='gray') 
print(f'Dice number=',img0_counter)   


#print(len(circle_img0))
```
