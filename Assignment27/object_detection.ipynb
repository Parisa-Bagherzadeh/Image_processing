```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img=cv2.imread('hat.jpg',0)

plt.imshow(img,cmap='gray')
```

```python
img=255-img
plt.imshow(img,cmap='gray')
```

```python
_,img=cv2.threshold(img,95,255,cv2.THRESH_BINARY)
plt.imshow(img,cmap='gray')
```

```python
height,width=img.shape
contour_index=[]
for i in range(height):
    for j in range(width):
        if (img[i,j]!=0):
            if((img[i,j+1]==0) or (img[i,j-1]==0) or (img[i+1,j]==0) or (img[i-1,j]==0)):
                contour_index.append([(j,i)])
points=np.array(contour_index)
#print(points)
               



```

```python


cv2.drawContours(img,points,-1,(60,25,40),10)
plt.imshow(img,cmap='gray')
```
