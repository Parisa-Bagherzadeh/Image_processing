```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img_logo=np.ones((600,1400,3),dtype='uint8')*87
plt.imshow(img_logo)
```

```python
img_logo[200:300,200:300]=(242,80,34)
img_logo[310:410,200:300]=(0,164,239)
img_logo[200:300,310:410]=(127,186,0)
img_logo[310:410,310:410]=(255,185,0)
cv2.putText(img_logo,'Microsoft',(450,350),cv2.FONT_HERSHEY_SIMPLEX,5,(255,255,255),15)
plt.imshow(img_logo)
```

```python
logo_result=cv2.cvtColor(img_logo,cv2.COLOR_BGR2RGB)
cv2.imwrite('logo_result.jpg',logo_result)
```
