```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_background=cv2.imread('space2.jpg')
img_background=cv2.cvtColor(img_background,cv2.COLOR_BGR2RGB)
plt.imshow(img_background)
```

```python
img=cv2.imread('image.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img_HSV=cv2.cvtColor(img,cv2.COLOR_RGB2HSV)
H,S,V=cv2.split(img_HSV)
```

```python
rows,cols=H.shape
img_background=cv2.resize(img_background,(cols,rows))
```

```python
for i in range(rows):
    for j in range(cols):
        if 35<H[i,j]<75 and S[i,j]>60 and V[i,j]>60:
            img[i,j]=img_background[i,j]


```

```python
plt.imshow(img)
```

```python
img_result=cv2.cvtColor(img,cv2.COLOR_RGB2BGR)
cv2.imwrite('img_result.jpg',img_result)
```
